```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LogisticRegression, LinearRegression
import random
import scipy.stats as stats
import matplotlib.colors as mcolors

random.seed(1249012780)
```

```python
mu1 = np.array([1,2])
mu2 = np.array([3,4])

Sigma = 1.5*np.array([[1,0],[0,1]])
X1 = np.random.multivariate_normal(mu1, Sigma, 100)
X2 = np.random.multivariate_normal(mu2, Sigma, 100)

plt.figure(figsize=(5,5))
plt.scatter(X1[:,0],X1[:,1],marker='.')
plt.scatter(X2[:,0],X2[:,1],marker='.')

plt.axis('equal')


a = plt.axis()
```

```python
r = 0.5
fig, ax = plt.subplots(figsize=(5,5)) # note we must use plt.subplots, not plt.subplot

for i in range(4):
    ax.add_patch(plt.Circle(mu1, (i+1)*r, color='blue',fill=False))
    ax.add_patch(plt.Circle(mu2, (i+1)*r, color='orange',fill=False))

plt.scatter(X1[:,0],X1[:,1],marker='.')
plt.scatter(X2[:,0],X2[:,1],marker='.')

plt.xlim(0,10)
plt.ylim(0,10)

plt.axis('equal')

plt.axis(a)
```

```python

sigma1 = lambda x :1/(1+np.exp(-x))
sigma2 = lambda x :1/(1+np.exp(-2*x))
             
plt.figure(figsize=(10,4))

x = np.linspace(-10,10,300)
plt.plot(x, sigma1(x),linewidth=3,color='magenta')
plt.plot(x, sigma2(x),linewidth=3,color='royalblue')

xbar = x[170]
plt.scatter(xbar,1,color='black',marker='o')
plt.vlines(xbar+0.06,0,sigma1(xbar),linewidth=3,color='magenta')
plt.vlines(xbar-0.06,0,sigma2(xbar),linewidth=3,color='royalblue')


xbar = x[120]
plt.scatter(xbar,0,color='red',marker='o')
plt.vlines(xbar+0.06,sigma1(xbar),1,linewidth=3,color='magenta')
plt.vlines(xbar-0.06,sigma2(xbar),1,linewidth=3,color='royalblue')



plt.hlines([0,1],-10,10, color='black', linestyles='--', linewidth=0.5)

plt.ylim([-0.2,1.2])
plt.xlim([-7.5,7.5])
```
