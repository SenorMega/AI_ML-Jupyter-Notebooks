# The Central Limit Theorem

The *Central Limit Theorem* states that if we draw a simple random sample of size $n$ from any population with mean $\mu$ and standard deviation $\sigma$, if $n$ is large the distribuion of the sample means is approximately normal of the form

$$N(\mu, \frac{\sigma}{n})$$

This activity is meant to offer an example of the central limit theorem in action, and the effect of using different sample sizes in forming the distribution of sample means.

### Taking Samples

Pandas DataFrame's have built in methods for taking random samples with replacement.  The `.sample()` method is demonstrated below.

```python
baseball_salaries['salary'].sample(4)
```

### Sample Means

Now, use this method to repeatedly take samples and build a list of sample means.  As a starting place let us take 1000 samples of size 5 and draw the histogram of the sample means.  What is the mean and standard deviation? 

```python
list_of_ns = [5,10,15,25,35,50,100,250,500,750,1000,1250,1500,1750,2000,2500,3000]
list_of_sample_means = []
list_of_sample_stds = []

#Find actual mean of dataset, for comparison
real_mean = np.mean(baseball_salaries['salary'])#.mean()
print('The actual mean is: ')
print(real_mean)

#empty list for sample means
sample_means = []

for sample_size in list_of_ns:

    for i in range(1000):
        #your code below
        
        #Version WITHOUT replacement of the random samples
        #Max sample size here is the size of the data set, so adjust list_of_ns
        #samples = baseball_salaries['salary'].sample(sample_size)

        #Version WITH replacement of the random samples
        #Any sample size
        samples = baseball_salaries['salary'].sample(sample_size, replace=True)
        
        sample_mean = np.mean(samples)
        sample_means.append(sample_mean)

    
    mean_of_sample_means = np.mean(sample_means)
    #print(mean_of_sample_means)
    list_of_sample_means.append(mean_of_sample_means)

    std_of_sample_means = np.std(sample_means)
    #print(std_of_sample_means)
    list_of_sample_stds.append(std_of_sample_means)


#print(list_of_sample_means)
#print(list_of_sample_stds)
plt.plot(list_of_ns, list_of_sample_means, 'o-', color='orange', label='Mean')
plt.xlabel('sample size')
plt.ylabel('mean')
plt.legend()
plt.title('Mean vs Sample Size')
plt.show()

plt.plot(list_of_ns, list_of_sample_stds, 'o-', color='blue', label='Standard Deviation')
plt.xlabel('sample size')
plt.ylabel('standard deviation')
plt.legend()
plt.title('Standard Deviation vs Sample Size')
plt.show()
```
