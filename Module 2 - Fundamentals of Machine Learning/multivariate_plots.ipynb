```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import multivariate_normal
```

```python
penguins = pd.read_csv('penguins.csv')

penguins = penguins[['Species','Culmen Length (mm)','Culmen Depth (mm)']]
# ,'Flipper Length (mm)','Body Mass (g)','Sex'
penguins.rename(columns={
 'Culmen Length (mm)':'Bill length',
 'Culmen Depth (mm)':'Bill depth',
#  'Flipper Length (mm)':'Flipper length',
#  'Body Mass (g)':'Body mass'
},inplace=True )

penguins['Species'].replace(
    {'Chinstrap penguin (Pygoscelis antarctica)':'Chinstrap',
     'Gentoo penguin (Pygoscelis papua)':'Gentoo',
     'Adelie Penguin (Pygoscelis adeliae)':'Adelie'},inplace=True)

# penguins['Sex'].replace({'FEMALE':'F','MALE':'M'},inplace=True)

# drop all rows with nan 
penguins.dropna(inplace=True)

# drop single row with sex='.'
# penguins = penguins[penguins['Sex'].isin({'F','M'})]

chinstrap = penguins[penguins['Species']=='Chinstrap']
gentoo = penguins[penguins['Species']=='Gentoo']
adelie = penguins[penguins['Species']=='Adelie']
```
