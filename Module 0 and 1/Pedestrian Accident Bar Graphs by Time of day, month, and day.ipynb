```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/sspillane/Desktop/Python Review Jupyter Notebooks/Module 0 and 1/data/ped_crashes.csv')
df_2018filter = df[ df['Crash Year']==2018 ]
#df_2018filter

#df.plot(kind='scatter', x='Worst Injury in Crash', y='Speed Limit at Crash Site')
```

```python
freq = df_2018filter['Time of Day'].value_counts()
freq.plot(kind='bar', title='Frequency Count by Time of Day', xlabel='Time of Day', ylabel='Count')
```

```python
freq2 = df_2018filter['Day of Week'].value_counts()
freq2.plot(kind='bar', title='Pedestrian Accident Frequency by Day of Week (2018)', xlabel='Day of Week', ylabel='Count')
```

```python
freq3 = df_2018filter['Crash Month'].value_counts()
freq3.plot(kind='bar', title='Pedestrian Accident Frequency by Month (2018)', xlabel='Count', ylabel='Month')
```

```python
freq4 = df_2018filter['Speed Limit at Crash Site'].value_counts()
freq4.plot(kind='bar')
```
