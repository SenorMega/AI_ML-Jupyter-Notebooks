Converting checking_status and savings_status to numeric buckets

````python
# Checking status are bucketed into -100, 0, 100 and 200

def checking_bucketing(tkn):
    if tkn == "\'<0\'":
        return '-100'
    elif tkn == "\'no checking\'":
        return '0'
    elif tkn =="\'0<=X<200\'":
        return '100'
    elif tkn == "\'>=200\'" :
        return '200'
    else:
        return '9999'

# Savings status are bucketed into -100, 0, 250, 750 and 1000

def savings_bucketing(tkn):
    if tkn == "\'<100\'":
        return '-100'
    elif tkn == "\'no known savings\'":
        return '0'
    elif tkn == "\'100<=X<500\'":
        return '250'
    elif tkn == "\'500<=X<1000\'" :
        return '750'
    elif tkn == "\'>=1000\'" :
        return '1000'
    else:
        return '9999'
    


df['checking_bucket'] = df['checking_status'].apply(lambda x : checking_bucketing(x))
df['savings_bucket'] = df['savings_status'].apply(lambda x : savings_bucketing(x))
````

````python
df['checking_bucket'] = pd.to_numeric(df['checking_bucket'])
df['savings_bucket'] = pd.to_numeric(df['savings_bucket'])
````

````python
# Importing visualization libraries
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
savings_plot = sns.boxplot(df, y='savings_bucket', x='credit_history')
savings_plot.set_xticklabels(savings_plot.get_xticklabels(),rotation=30)
````

````python
df['employment'].value_counts()
````

````python
# Bucketing employment

def employment_bucketing(tkn):
    if tkn == "\'<1\'":
        return '0'
    elif tkn == "unemployed":
        return '0'
    elif tkn == "\'1<=X<4\'":
        return '4'
    elif tkn == "\'4<=X<7\'" :
        return '7'
    elif tkn == "\'>=7\'" :
        return '10'
    else:
        return '9999'
````

````python
df['employment_bucket'] = df['employment'].apply(lambda x : employment_bucketing(x))
````

````python
df['employment_bucket'] = pd.to_numeric(df['employment_bucket'])
````

````python
employment_plot = sns.boxplot(df, y='employment_bucket', x='credit_history')
employment_plot.set_xticklabels(employment_plot.get_xticklabels(),rotation=30)
````
