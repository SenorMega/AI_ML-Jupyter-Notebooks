```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, uniform
```

```python
def samplemean_vs_Gauss(X,n,m=5000): 
    # n ... sample size
    # m ... number of draws for the histogram
    Xstderr = X.std()/np.sqrt(n)
    data = [ np.sqrt(n)*(X.rvs(size=n).mean()-X.mean())/X.std() for i in range(m)]
    plt.hist(data,density=True,bins=30)
    x = np.linspace(-3,3)
    stdN = norm.pdf(x)
    plt.plot(x,stdN,linewidth=3)
```

```python
U = uniform(loc=10, scale=5)
samplemean_vs_Gauss(U,n=50)
```

```python
B = bernoulli(p=0.5)
samplemean_vs_Gauss(B,n=50)
```
