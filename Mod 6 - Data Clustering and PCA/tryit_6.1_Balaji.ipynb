##### Plotting in 3D

Below, example plots are made with both `matplotlib` and with `plotly` to demonstrate how to construct three dimensional plots.  

`matplotlib`

In `matplotlib` an `axes` object is created where the projection is specified as '3d'.  Then, the `axes` are use to plot a 3D object on.  Below, a scatter plot is built, but there are many 3D objects that can be created in addition to points.  For more information see the documentation on 3D plotting [here](https://matplotlib.org/stable/api/toolkits/mplot3d.html?highlight=3d).


`plotly`

Use the `scatter_3d` function directly.

```python
#Standarize the dataset
U = default.mean()
Sigma = default.std()
default_ = (default - U)/Sigma   
 
#Apply PCA function
pca = PCA(n_components=3) 
default_pca = pca.fit_transform(default_)

#Plot scaled data on a 3D plot
plt.figure(figsize = (8,6)) 
ax = plt.axes(projection = '3d')
ax.scatter3D(default_pca[:, 0], default_pca[:, 1], default_pca[:, 2], c = default['default.payment.next.month'], alpha = 0.4)
ax.set_xlabel('Column 0', labelpad = 20)
ax.set_ylabel('Column 1', labelpad = 20)
ax.set_zlabel('Default', labelpad = 20)
plt.title('3D representation of decomposed features')
plt.tight_layout();
```

```python
# Plot the transformed data on a 2D plot
plt.scatter(default_pca[:, 0], default_pca[:, 1], c=default['default.payment.next.month'])
plt.title('2D representation of decomposed features')
plt.xlabel('Column 0')
plt.ylabel('Column 1')
plt.show()
```
